```python
# Ornek 4:
#Kullanıcının tuttuğu 1 - 10 arasi sayıyı tahmin eden oyun yapiniz.

kullanici_sayi = 4
tahmin_sayisi = 0
 
while True:
    tahmin_sayisi += 1
    sayi=int(input("1 ile 10 arasında değer girin (0 Çıkış):"))
    if(sayi==0):
        print("Oyunu İptal Ettiniz")
        break
    elif sayi < kullanici_sayi:
        print("Daha Yüksek Bir Sayı Girin.")
        
    elif sayi > kullanici_sayi:
        print("Daha Düşük Bir Sayı Girin.")
        
    else:
        print("Tebrikler! {}. Tahmininizde Bildiniz! Tutulan sayi {}!".format(tahmin_sayisi, kullanici_sayi))
        break 
```

```python
# Bonus
#Kullanıcının girmiş olduğu iki sayı arasindaki (bu sayilar dahil) tek ve çift sayıların toplamını ayrı ayrı bulup sonucu gösteriniz.

sayi_min = int(input('Kucuk Sayıyı Girin : '))
sayi_max = int(input('Buyuk Sayıyı Girin : '))
tek_toplam=0
cift_toplam=0
for i in range(sayi_min,sayi_max + 1):
    if(i%2==0):
        cift_toplam += i
    else:
        tek_toplam += i

print("Tek Sayıların Toplamı : {0}".format(tek_toplam))
print("Çift Sayıların Toplamı : {0}".format(cift_toplam))
```

```python
sayi_min = int(input('Kucuk Sayıyı Girin : '))
sayi_max = int(input('Buyuk Sayıyı Girin : '))
tek_toplam=0
cift_toplam=0

if sayi_max > sayi_min:
    for i in range(sayi_min,sayi_max + 1):
        if(i%2==0):
            cift_toplam += i
        else:
            tek_toplam += i
    print("Tek Sayıların Toplamı : {0}".format(tek_toplam))
    print("Çift Sayıların Toplamı : {0}".format(cift_toplam))
else:
    print("Yaw sen git kod yazmayi birak. Once max ve min ne onu ogren.")
```
