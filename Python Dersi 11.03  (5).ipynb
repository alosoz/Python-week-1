##While:

```python
i = 1
while i < 6:
    print(i)
    i += 2
```

```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1 
```

##For:

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
```

```python
for x in "banana":
    print(x)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
    if x == "banana":
        break
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        break
    print(x)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        continue
    print(x)
```

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
    for y in fruits:
        print(x, y) 
```
